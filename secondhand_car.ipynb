```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('train.csv')
```

```python
test = pd.read_csv('test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

train.fillna({'condition': 'unknown', 'cylinders': 'unknown', 'fuel': 'unknown',
                   'title_status': 'unknown', 'transmission': 'unknown', 'drive': 'unknown',
                   'size': 'unknown', 'type': 'unknown', 'paint_color': 'unknown', 'state': 'unknown'}, inplace=True)
test.fillna({'condition': 'unknown', 'cylinders': 'unknown', 'fuel': 'unknown',
                  'title_status': 'unknown', 'transmission': 'unknown', 'drive': 'unknown',
                  'size': 'unknown', 'type': 'unknown', 'paint_color': 'unknown', 'state': 'unknown'}, inplace=True)
```

```python

label_encoders = {}
for column in ['region', 'manufacturer', 'condition', 'cylinders', 'fuel', 'title_status',
               'transmission', 'drive', 'size', 'type', 'paint_color', 'state']:
    le = LabelEncoder()
    train[column] = le.fit_transform(train[column])
    test[column] = le.transform(test[column])
    label_encoders[column] = le
```

```python
X_train = train.drop(columns=['id', 'price'])
y_train = train['price']
X_test = test.drop(columns=['id'])
```

```python
X_train.head(), y_train.head()
```

```python
import xgboost as xgb

xgb_model = xgb.XGBRegressor(objective ='reg:squarederror', random_state=42)

xgb_model.fit(X_train, y_train)

train_score = xgb_model.score(X_train, y_train)
train_score
```

```python

price_predictions = xgb_model.predict(X_test)

submit_df = pd.DataFrame({'id': test['id'], 'price': price_predictions})

submit_file_path = "/content/drive/MyDrive/secondhand_car/data/submit_xgb.csv"
submit_df.to_csv(submit_file_path, index=False,header=False)

submit_df.head(), submit_file_path
```
